```python
import pandas as pd
import pickle
import re
# Load and preprocess dataset
data = pd.read_csv("/content/drive/MyDrive/gpt-4.csv", engine='python', on_bad_lines='skip')

# Define symptoms and diseases lists
symptoms = ['Bleeding', 'Chills', 'Cough', 'Diarrhea', 'Dizziness', 'Dyspnea', 'Fatigue', 'Fever', 'Headache',
            'Loss of appetite', 'Nausea', 'Numbness', 'Pain', 'Rash', 'Seizure', 'Shortness of breath', 'Sweating',
            'Swelling', 'Tingling', 'Vomiting', 'Weakness', 'Stress']

diseases = ['ARDS', 'Arthritis', 'Asthma', 'Cancer', 'Carcinoma', 'Condition', 'Diabetes', 'Disease', 'Disorder',
            'Failure', 'Hypertension', 'Illness', 'Infection', 'Lesion', 'Metastases', 'Pneumonia', 'Stroke',
            'Syndrome', 'Tumor']

# Function to extract gender
def extract_gender(text):
    if re.search(r'\b(she|her|female|woman|girl|miss)\b', text, re.IGNORECASE):
        return 'Female'
    elif re.search(r'\b(he|his|male|man|boy|mr)\b', text, re.IGNORECASE):
        return 'Male'
    return 'Unknown'

# Function to extract age
def extract_age(text):
    text = re.sub(r'[^a-zA-Z0-9\s]', '', text)
    match = re.search(r'(\d+)\s*year', text, re.IGNORECASE)
    if match:
        return int(match.group(1))
    return None

# Apply age and gender extraction
data['age'] = data['data'].apply(extract_age)
data['gender'] = data['data'].apply(extract_gender)

# Sample data and extract symptoms and diseases
data = data.sample(n=min(3000, len(data)))
data['symptoms'] = data['data'].apply(lambda x: [s for s in symptoms if s.lower() in x.lower()])
data['diseases'] = data['data'].apply(lambda x: [d for d in diseases if d.lower() in x.lower()])
data['serial_number'] = range(1, len(data) + 1)

# Save preprocessed data to a pickle file
with open('data.pkl', 'wb') as file:
    pickle.dump(data, file)

print("Data has been saved to data.pkl")
```
